```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread(r'C:\Users\Yunita Elisabet\OneDrive\Dokumen\semester 4\citra digital\chess.jpg',0)          # queryImage
img2 = cv2.imread(r'C:\Users\Yunita Elisabet\OneDrive\Dokumen\semester 4\citra digital\dga.jpeg',0)
```

```python
orb = cv2.ORB_create()

```

```python
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
matches = bf.match(des1,des2)
```

```python
matches = sorted(matches, key = lambda x:x.distance)
```

```python
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:30],None, flags=2)
```

```python
plt.imshow(img3),plt.show()
```
